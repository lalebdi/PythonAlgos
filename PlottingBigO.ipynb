```python
def sum1(n):
    final_sum = 0
    for i in range(n+1):
        final_sum += i
    return final_sum
```

```python
sum1(5)
```

```python
def sum2(n):
    return (n*(n+1))/2
```

```python
sum2(5)
```

```python
%timeit sum1(100)
```

```python
%timeit sum2(100)
```

```python
def bigO(n):
    return 45*n**3 + 20 * n**2 + 19
```

```python
from math import log
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('bmh')

# Set up time comparisons
n = np.linspace(1, 10)
labels = ['Constant', 'Logarithmic', 'Linear', 'Log Linear', 'Quadratic', 'Cubic', 'Exponential']
big_o = [np.ones(n.shape), np.log(n), n, n*np.log(n), n**2, n**3, 2**n]

# Plot Setup
plt.figure(figsize=(12,10))
plt.ylim(0, 50)

for i in range(len(big_o)):
    plt.plot(n, big_o[i], label = labels[i])
    
plt.legend(loc=0)
plt.ylabel('Relative Runtime')
plt.xlabel('n')
```
